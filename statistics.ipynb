```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats 
```

```python
df = pd.read_csv('cleaned_fitness_data.csv')
df.head()
```

```python
population_mean = df['duration_min'].mean()
print(f"The mean: {population_mean}")
```

```python
sample_mean = df['duration_min'].head(100).mean()
print(f"The first 100 mean: {sample_mean}")
```

```python
population_variance = df['duration_min'].var(ddof = 0)
print(f"Population variance: {population_variance}")
```

```python
sample_variance = df['duration_min'].head(100).var( ddof = 1)
print(f" Sample Variance: {sample_variance}")
```

```python
try:
    sample = [np.mean(df['duration_min'].sample(14)) for _ in range (100)]
except Exception as e:
    print(e)
```

```python
def histogram(data):
    try:
        plt.hist(data, bins = 10, edgecolor = 'black')
        plt.xlabel('Duration in minutes')
        plt.ylabel('Frequency')
    except Exception as e:
        print(e)



histogram(sample)        
```

```python
mean = np.mean(df['duration_min'].head(100))
standard_error = stats.sem(df['duration_min'].head(100))
degree_of_freedom = len(df['duration_min'].head(100))-1
confidence_level = 0.95
confidence_interval = stats.t.interval(confidence_level, degree_of_freedom, mean, standard_error)
print(f"95% confidence level: {confidence_interval}")
```
